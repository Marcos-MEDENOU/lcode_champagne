```python
# Convertir les colonnes 'date' et 'heure' en type 'datetime'
fibres_leads['date'] = pd.to_datetime(fibres_leads['date'])
fibres_leads['heure'] = pd.to_datetime(fibres_leads['heure'], format='%H:%M:%S').dt.time
```

```python
# Grouper les données par jour
grouped = fibres_leads.groupby('date')

# Créer une liste vide pour stocker les données triées
sorted_data = []
```

```python
# Pour chaque groupe de données, trier les données par heure et par date
for date, group in grouped:
    sorted_group = group.sort_values(['heure', 'date'], ascending=[True, True])
    sorted_data.append(sorted_group)
```

```python
# Concaténer les données triées en un seul DataFrame
fibres_leads = pd.concat(sorted_data, ignore_index=True)
```

```python
# Concaténer les colonnes date et heure en une seule colonne datetime
fibres_leads["datetime"] = fibres_leads["date"].astype(str) + " " + fibres_leads["heure"].astype(str)
```

```python
# Ajouter des colonnes pour chaque jour de la semaine
#['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
#jours_semaine = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
jours_semaine = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
for jour in jours_semaine:
    fibres_leads[jour] = 0  # initialiser la colonne à 0
# Parcourir chaque ligne du DataFrame
for index, row in fibres_leads.iterrows():
    # Extraire la date de la ligne
    date = row['date']
    # Extraire le jour de la semaine correspondant à la date
    nom_jour = date.strftime('%A')
    # Ajouter la valeur de la colonne 'jour' à la colonne correspondante
   # fibres_leads.loc[index, nom_jour] += row['Jour']
    if pd.isnull(fibres_leads.loc[index, nom_jour]):
        fibres_leads.loc[index, nom_jour] = row['Jour']
    else:
        fibres_leads.loc[index, nom_jour] += row['Jour']
```
